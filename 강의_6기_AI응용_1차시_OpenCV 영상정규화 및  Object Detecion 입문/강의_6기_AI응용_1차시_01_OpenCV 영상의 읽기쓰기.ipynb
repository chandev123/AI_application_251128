**3차원이 뭐예요?** 📦

**차원별 의미:**
1. **1차원 (높이)**: 이미지의 세로 (위↔아래)
2. **2차원 (너비)**: 이미지의 가로 (좌↔우)
3. **3차원 (채널)**: 색상 정보 (R, G, B 3가지)

**비유로 이해하기:**
- 아파트 주소를 생각해보세요
  - 1차원: 101호, 102호, 103호... (일렬)
  - 2차원: 1층 101호, 2층 201호... (평면)
  - 3차원: A동 1층 101호, B동 2층 201호... (입체)

**흑백 vs 컬러:**
```
흑백 사진 = 2차원 [높이][너비]
컬러 사진 = 3차원 [높이][너비][RGB 3개]
```

uint8 의미는?

- unsigned : 부호(+,-) 없음 (음수 없음, 0 이상만)
- 8 : 2^8 = 256 (즉, 256가지 값으로 표현)
- 256 값 >> 0-255 범위 갖는다
- 0: 완전 어두워(검정), 255: 완전 밝아 (하양)

논리연산
- AND(교집합): 둘 다 True(1)
- OR(합집합): 둘 다 False(0) / 하나만 True여도 True
- XOR(-합집합, 합집합 역수): 하나가 True면 다른 하나가 True여야 True
- NOT(원래 값 뒤집기)

````python
#도형 생성
#np.zeros(크기, 데이터타입) = 0이라는 값으로 채워진 '크기' 만큼의 데이터를 만듦
np.zeros((200, 200), dtype=np.uint8)
````

````python
# 도화지
# np.zeros(사이즈,타입)
img1 = np.zeros((200, 200, 3), dtype=np.uint8)
img2 = np.zeros((200, 200, 3), dtype=np.uint8)

# cv2.rectangle()
# cv2.rectangle(바탕 이미지, 시작점, 끝점, 색, [두께])
# 색은 채널 기준으로 표현
cv2.rectangle(img1, (50, 50), (150, 150), (125, 125, 30), 2)
````

````python
cv2.rectangle(img1, (50, 50), (150, 150), (125, 125, 30), -1) # -1 채움
````

````python
# (image, 중심점, 반지름, 색깔, 두께)
cv2.circle(img2, (100, 100), 70, (40, 170, 70), 2)
````

````python
cv2.circle(img2, (100, 100), 70, (40, 170, 70), -1)
````

````python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1,2,figsize=(6,3))
axes[0].imshow(img1)
axes[0].set_title('rectangle')
axes[1].imshow(img2)
axes[1].set_title('circle')
plt.show()
````

````python
# AND 연산(교집합) : 겹치는 부분만 그림에 넣는다
bit_and = cv2.bitwise_and(img1, img2)
cv2.add(bit_and, 100)
````

````python
bit_and = cv2.add(img1, img2)
fig, axes = plt.subplots(1,2, figsize=(6,3))
axes[0].imshow(img1, cmap='gray') # cmap : 컬러맵
axes[0].set_title('rectangle')
axes[1].imshow(bit_and) # cmap : 컬러맵
axes[1].set_title('rectangle_bit')

plt.show()
````

````python
bit_and = cv2.add(img1, img2)
fig, axes = plt.subplots(1,2, figsize=(6,3))
axes[0].imshow(img1, cmap='gray') # cmap : 컬러맵
axes[0].set_title('rectangle')
axes[1].imshow(bit_and, cmap='gray') # cmap : 컬러맵
axes[1].set_title('rectangle_bit')

plt.show()
````

````python
#AND연산(교집합) -> 겹치는 부분만 그림에 남는다.
bit_and = cv2.bitwise_and(img1, img2)
bit_and = cv2.add(bit_and, 100)
fig, axes = plt.subplots(1, 2, figsize=(6,3))

axes[0].imshow(img1, cmap='gray')
axes[0].set_title('rectangle')

axes[1].imshow(bit_and, cmap='gray')
axes[1].set_title('changes')
plt.show()
````

````python
#OR연산(합집합) -> 두 쪽 중 어느 하나라도 포함되는 픽셀은 표현
img1 = np.zeros((200, 200, 3), dtype=np.uint8)
img2 = np.zeros((200, 200, 3), dtype=np.uint8)

cv2.rectangle(img1, (50, 50), (150, 150), (125, 125, 30), -1)
cv2.circle(img2, (100,100), 70, (40,170,70), -1)

bit_or2 = cv2.bitwise_or(img1, img2)
fig, axes = plt.subplots(1, 2, figsize=(6,3))

axes[0].imshow(img1, cmap='gray')
axes[0].set_title('rectangle')

axes[1].imshow(bit_or2, cmap='gray')
axes[1].set_title('changes')
plt.show()
````

````python
#XOR
img1 = np.zeros((200, 200, 3), dtype=np.uint8)
img2 = np.zeros((200, 200, 3), dtype=np.uint8)

cv2.rectangle(img1, (50, 50), (150, 150), (125, 125, 30), -1)
cv2.circle(img2, (100,100), 70, (40,170,70), -1)

bit_xor = cv2.bitwise_xor(img1, img2)
fig, axes = plt.subplots(1, 2, figsize=(6,3))

axes[0].imshow(img1, cmap='gray')
axes[0].set_title('rectangle')

axes[1].imshow(bit_xor, cmap='gray')
axes[1].set_title('changes')
plt.tight_layout()
plt.show()
````

````python
#NOT
img1 = np.zeros((200, 200, 3), dtype=np.uint8)
img2 = np.zeros((200, 200, 3), dtype=np.uint8)

cv2.rectangle(img1, (50, 50), (150, 150), (125, 125, 30), -1)

bit_not = cv2.bitwise_not(img1)
fig, axes = plt.subplots(1, 2, figsize=(6,3))

axes[0].imshow(img1, cmap='gray')
axes[0].set_title('rectangle')

axes[1].imshow(bit_not, cmap='gray')
axes[1].set_title('changes')
plt.tight_layout()
plt.show()
````
